```python
def ingestion_func(schema:tuple,load_url:str):
    df = (
        spark.read.format("csv")
        .schema(schema=schema)
        .option("header","true")
        .load(load_url)
    )

    return df
```

```python
def save_to_table(df,table_name:str,mode: str):
    try:
        df.write.mode(mode).saveAsTable(table_name)
        return "Okay"
    except Exception as e:
        print(f"Error: {e}")
```
